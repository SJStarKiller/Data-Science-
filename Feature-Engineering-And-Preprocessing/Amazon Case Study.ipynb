```python
from sklearn.feature_extraction import text
```

```python
tf=text.TfidfVectorizer(input=data,ngram_range=(1,2),stop_words='english')
mat=tf.fit_transform(data)
pd.DataFrame(mat.toarray(),columns=tf.get_feature_names())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_unigram=cosine_similarity(mat)
```

```python
sim_unigram[0].argsort()[-5:-1]
```

```python
df['review_detail'].loc[sim_unigram[0].argsort()[-5:-1]]
```

```python
df['review_detail'].str.replace('_',' ').str.upper().str.strip()[5]
```

```python
def get(x):
    return ','.join(df['review_detail'].loc[x.argsort()[-5:-1]])
df['cosine']=[get(x) for x in sim_unigram]
```
