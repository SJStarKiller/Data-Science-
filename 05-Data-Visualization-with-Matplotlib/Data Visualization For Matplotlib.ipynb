## Uni-Variate Plots
#### 1 Histogram

```python
plt.hist(data.Income)
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.hist(data.Income,bins=10,color='blue',alpha=0.7,orientation='horizontal');
```

```python
plt.figure(figsize=(15,7))
plt.hist(data.Income,bins=10,color='blue',alpha=0.7);
plt.xlabel('Income')
plt.ylabel('Frequency of no. of employees')
plt.title('Histogram Income Plot Demo')
plt.savefig('Histogram_plot.jpg')
```

#### 2 Density Plot

```python
data.Income.plot(kind='density',ls='--')
```

```python
plt.figure(figsize=(15,9))
data.Income.plot(kind='density',ls='--',secondary_y=True);
plt.xlabel('Income')
plt.ylabel('Frequency of no. of employees')
plt.title('Density Income Plot Demo')
plt.tight_layout()
plt.show()
```

## Hist with density

```python
plt.figure(figsize=(15,9))
ax=data.Income.plot(kind='hist',color='white',ylim=400)
ax2=data.Income.plot(kind='density',color='blue',ls='--',secondary_y=True)
ax.set_ylabel('No. of emp')
ax2.set_ylabel('Density')
plt.tight_layout()
plt.show()
```
